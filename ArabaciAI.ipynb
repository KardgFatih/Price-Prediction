```python
cleanDataFrame = cleanDataFrame.drop("transmission", axis = 1)
y = cleanDataFrame["price"].values
x = cleanDataFrame.drop("price", axis= 1).values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.33, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
xTrain = scaler.fit_transform(xTrain)
xTest = scaler.fit_transform(xTest)

len(xTrain) 
#len(yTrain) 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))
model.compile(optimizer = "adam", loss="mse")
```

```python
model.fit(x = xTrain , y=yTrain, validation_data=(xTest,yTest),batch_size=250,epochs = 300)
```

```python
losingData = pd.DataFrame(model.history.history)
losingData.head()
```

```python
losingData.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
predictionList = model.predict(xTest)
```
